```python
# Function to count outliers

num_cols = ['age', 'bmi', 'children', 'expenses']

def count_outliers_iqr(df_clean, column):
    Q1 = df_clean[column].quantile(0.25)
    Q3 = df_clean[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df_clean[(df_clean[column] < lower) | (df_clean[column] > upper)].shape[0]

for col in num_cols:
    outliers = count_outliers_iqr(df_clean, col)
    print(f"{col}: {outliers} outliers")
```

```python
# Keep the expenses outliers because they're not errors. These high-cost entries (likely due to smoking, obesity, etc.) are exactly what the model needs to learn to predict well
# Cap the bmi outliers because very high bmi values (like 50+) can dominate linear models.
# So, capping BMI only

def cap_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[column] = df[column].clip(lower, upper)
    return df

df_clean = cap_outliers_iqr(df_clean, 'bmi')
```

# üß† 3. Feature Engineering

```python
# üì¶ Imports

from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder
```

```python
# Separate features and target

X = df_clean.drop('expenses', axis=1)
y = df_clean['expenses']
```

```python
# Identify column types

num_cols  = X.select_dtypes(include=["int64", "float64"]).columns.tolist()
cat_cols  = X.select_dtypes(exclude=["int64", "float64"]).columns.tolist()
```

```python
# Split categorical into **binary** vs **multi-class** for label vs one-hot
binary_cols = [c for c in cat_cols if X[c].nunique() == 2]
ohe_cols    = [c for c in cat_cols if X[c].nunique() > 2]
```

```python
# Pre-processing transformers

# Numeric features pipeline: Just scaling
numeric_pipe = Pipeline([
    ("scale", StandardScaler())
])

# Binary categorical features: Just label encoding (OrdinalEncoder)
binary_pipe = Pipeline([
    ("label", OrdinalEncoder())
])

# Multiclass categorical features: One-hot encoding
ohe_pipe = Pipeline([
    ("ohe", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
])

# Combine all in ColumnTransformer
preprocess = ColumnTransformer([
    ("num",  numeric_pipe,  num_cols),
    ("bin",  binary_pipe,   binary_cols),
    ("ohe",  ohe_pipe,      ohe_cols)
])
```

```python
# üì¶ Imports
from sklearn.model_selection import train_test_split

# Train / Val / Test split (70 / 15 / 15)

# Hold-out final test
X_temp, X_test, y_temp, y_test = train_test_split(
    X, y, test_size=0.15, random_state=42)

# Validation from remaining 85 %
val_size = 0.15 / 0.85          # ‚âà0.176
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=val_size, random_state=42)

print(f"Shapes  ‚Üí  train:{X_train.shape}  val:{X_val.shape}  test:{X_test.shape}")
```

# ü§ñ 4. Model Building and Evaluation

```python
# üì¶ Imports
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.feature_selection import RFECV
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor

import warnings
warnings.filterwarnings("ignore")
```

```python
# Helper ‚Äì evaluation

def report(model, X_tr, y_tr, X_va, y_va, name):
    pred_val = model.predict(X_va)
    print(f"‚Äî‚Äî {name} ‚Äî‚Äî")
    print(f"Val  | MAE:{mean_absolute_error(y_va, pred_val):.2f} "
          f"RMSE:{np.sqrt(mean_squared_error(y_va, pred_val)):.2f} "
          f"R¬≤:{r2_score(y_va, pred_val):.4f}")
    pred_tr = model.predict(X_tr)
    print(f"Train| R¬≤ :{r2_score(y_tr, pred_tr):.4f}\n")
```

```python
#  Define model pipelines + search grids

models = {}

# --- A. Multiple Linear Regression  + RFECV ------------------
pipe_lr = Pipeline([
    ("prep", preprocess),
    ("rfe",  RFECV(estimator=LinearRegression(),
                   cv=5, scoring="neg_mean_absolute_error")),
    ("model", LinearRegression())
])
models["LinearRegression"] = {
    "pipeline": pipe_lr, "grid": {}
}

# --- B. Ridge  + RFECV ---------------------------------------
pipe_ridge = Pipeline([
    ("prep", preprocess),
    ("rfe",  RFECV(estimator=Ridge(),
                   cv=5, scoring="neg_mean_absolute_error")),
    ("model", Ridge())
])
models["Ridge"] = {
    "pipeline": pipe_ridge,
    "grid": {"model__alpha": [0.01, 0.1, 1, 10, 100]}
}

#--- C. SVR -----------------------------------------
pipe_svr = Pipeline([
    ("prep", preprocess),
    ("model", SVR()) # Removed RFECV for SVR
])
models["SVR"] = {
    "pipeline": pipe_svr,
    "grid": {
        "model__kernel": ["rbf", "linear"],
        "model__C": [1, 10, 100],
        "model__epsilon": [0.01, 0.1, 1]
    }
}

# --- D. Lasso (built-in feature selection) -------------------
pipe_lasso = Pipeline([
    ("prep", preprocess),
    ("model", Lasso(max_iter=5000))
])
models["Lasso"] = {
    "pipeline": pipe_lasso,
    "grid": {"model__alpha": [0.001, 0.01, 0.1, 1, 10]}
}

# --- E. Decision Tree ---------------------------------------
pipe_dt = Pipeline([
    ("prep", preprocess),
    ("model", DecisionTreeRegressor(random_state=42))
])
models["DecisionTree"] = {
    "pipeline": pipe_dt,
    "grid": {
        "model__max_depth": [None, 4, 6, 8, 10],
        "model__min_samples_split": [2, 5, 10],
        "model__min_samples_leaf": [1, 2, 4]
    }
}

# --- F. Random Forest ---------------------------------------
pipe_rf = Pipeline([
    ("prep", preprocess),
    ("model", RandomForestRegressor(n_estimators=100,random_state=42))
])
models["RandomForest"] = {
    "pipeline": pipe_rf,
    "grid": {
        "model__n_estimators": [300, 500, 700],
        "model__max_depth": [None, 6, 10, 14],
        "model__min_samples_leaf": [1, 2, 4]
    }
}

# --- G. Gradient Boosting -----------------------------------
pipe_gbr = Pipeline([
    ("prep", preprocess),
    ("model", GradientBoostingRegressor(n_estimators=100, learning_rate=0.1,random_state=42))
])
models["GBR"] = {
    "pipeline": pipe_gbr,
    "grid": {
        "model__n_estimators": [200, 400, 600],
        "model__learning_rate": [0.01, 0.05, 0.1],
        "model__max_depth": [2, 3, 4]
    }
}

# --- H. XGBoost ---------------------------------------------
pipe_xgb = Pipeline([
    ("prep", preprocess),
    ("model", XGBRegressor(
        n_estimators=100,
        learning_rate=0.1,
        objective="reg:squarederror",
        random_state=42,
        verbosity=0))
])
models["XGBoost"] = {
    "pipeline": pipe_xgb,
    "grid": {
        "model__n_estimators": [400, 600, 800],
        "model__learning_rate": [0.01, 0.05, 0.1],
        "model__max_depth": [3, 4, 5],
        "model__subsample": [0.8, 1.0],
        "model__colsample_bytree": [0.8, 1.0]
    }
}
```

```python
# Hyper-parameter search + validation performance

best_models, val_mae = {}, {}

for name, cfg in models.items():
    print(f"### Tuning {name}")
    pipe, grid = cfg["pipeline"], cfg["grid"]
    # decide grid vs random search
    combos = np.prod([len(v) for v in grid.values()]) if grid else 1
    if combos <= 40:
        search = GridSearchCV(pipe, grid,
                              cv=5, n_jobs=-1,
                              scoring="neg_mean_absolute_error")
    else:
        search = RandomizedSearchCV(pipe, grid, n_iter=50,
                                    cv=5, n_jobs=-1,
                                    random_state=42,
                                    scoring="neg_mean_absolute_error")
    search.fit(X_train, y_train)
    best = search.best_estimator_
    best_models[name] = best
    val_mae[name] = mean_absolute_error(y_val, best.predict(X_val))
    report(best, X_train, y_train, X_val, y_val, name)
```

```python
# --- Step 1: Initialize ---
best_model_name = None
best_model = None
best_r2 = float('-inf')

# --- Step 2: Loop through all trained models ---
for name, model in best_models.items():
    y_val_pred = model.predict(X_val)
    r2_val = r2_score(y_val, y_val_pred)

    print(f"{name}: R¬≤ (Val) = {round(r2_val, 4)}")

    if r2_val > best_r2:
        best_r2 = r2_val
        best_model = model
        best_model_name = name

# --- Step 3: Output the best model ---
print("\nüèÜ Best Model Based on R¬≤ (Val):")
print(f"Model: {best_model_name}")
print(f"R¬≤ Score: {round(best_r2, 4)}")

#  R¬≤ (coefficient of determination) directly tells you how much variation in the target variable (medical cost) is explained by the model.

# """We selected the best model based on R¬≤ on the validation set because it directly reflects how well the model explains the variation in medical costs.
#   While MAE and RMSE were used to support our analysis, R¬≤ provides a scale-independent and interpretable metric to compare model performance in terms of predictive power.
#   Our best model, XGBoost, achieved the highest R¬≤ of 0.8382, indicating strong generalization and accuracy for this regression task."""
```
